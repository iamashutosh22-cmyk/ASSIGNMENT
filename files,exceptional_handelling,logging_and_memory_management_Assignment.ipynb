# Q1 =What is the difference between interpreted and compiled languages

Compiled languages translate the entire source code into machine code before execution, while interpreted languages execute code line by line through an interpreter. In a compiled language like C or C++, the program is first converted into machine-readable instructions by a compiler, which makes execution faster since the computer runs the compiled binary directly; however, this requires a compilation step before running. In contrast, interpreted languages such as Python or JavaScript rely on an interpreter that reads and executes instructions one at a time, which makes them easier to debug and more flexible but generally slower in performance compared to compiled language



# Q2 = What is exception handling in Python


Exception handling in Python is a mechanism that lets you deal with runtime errors gracefully without crashing the program. It uses the try...except block, where code that might raise an error is placed inside the try block, and the except block catches and handles the error if it occurs. You can also use else to run code when no exception happens and finally to execute cleanup code regardless of whether an exception occurred. This approach allows developers to anticipate problems like invalid input, missing files, or division by zero, respond to them safely, and keep the program running smoothly


# Q3 = What is the purpose of the finally block in exception handling



The finally block in Python exception handling is used to define a section of code that will always execute regardless of whether an exception occurs or not, making it ideal for cleanup tasks such as closing files, releasing resources, or disconnecting from databases; this ensures that essential operations are performed even if an error interrupts the normal flow of the program.

# Q6 = What is the difference between import and from ... import in Python



In Python, the import statement brings in an entire module, meaning you must reference its functions or classes with the module name as a prefix (for example, import math requires you to call math.sqrt(16)), while the from ... import statement allows you to import specific attributes directly from a module so you can use them without the module prefix (for example, from math import sqrt lets you call sqrt(16) directly); the key difference is that import keeps the namespace organized by requiring explicit module references, whereas from ... import provides convenience and brevity but can risk name conflicts if multiple modules have functions or classes with the same name.

# Q9 = What is the difference between multithreading and multiprocessing



Multithreading and multiprocessing differ in how they achieve parallelism: multithreading uses multiple threads within a single process to share the same memory space, making it lightweight and efficient for tasks like I/O operations but limited by Python’s Global Interpreter Lock (GIL), which prevents true parallel execution of threads; multiprocessing, on the other hand, creates separate processes with independent memory spaces, allowing true parallel execution on multiple CPU cores, making it better suited for CPU-intensive tasks though it requires more resources and inter-process communication overhead.

# Q15 = How does Python's garbage collection system work



Python’s garbage collection system works by automatically managing memory through a combination of reference counting and cyclic garbage collection; every object keeps track of how many references point to it, and when this count drops to zero, the memory is immediately freed, while for objects involved in reference cycles that cannot be cleaned up by reference counting alone, Python’s cyclic garbage collector periodically detects and breaks these cycles to reclaim memory, ensuring efficient resource usage and preventing memory leaks without requiring manual intervention from the programmer.

# Q18 = What is the difference between os.fork() and multiprocessing in Python

# Q12 = What is the difference between file.read() and file.readline() in Python


The difference between file.read() and file.readline() in Python lies in how they retrieve data from a file:

file.read(): Reads the entire file content (or a specified number of characters if you pass an argument) into a single string. For example, file.read() without arguments will load the whole file at once, which is useful when you want all the data together but can be memory-heavy for large files.

file.readline(): Reads only one line from the file at a time, returning it as a string. Each call moves the file pointer to the next line, making it more memory-efficient and suitable for processing files line by line, especially large ones.

In short, file.read() is for bulk reading, while file.readline() is for sequential, line-by-line reading.

```python
# Q21 = Write a Python program that reads a file and prints the number of occurrences of a specific word


def count_word_occurrences(filename, word):
    try:
        with open(filename, "r") as f:
            content = f.read()
            # Split content into words and count matches
            words = content.split()
            count = words.count(word)
            print(f"The word '{word}' occurs {count} times in the file.")
    except FileNotFoundError:
        print("File not found.")

# Example usage
count_word_occurrences("example.txt", "Python")
```

```python
# Q22 = How can you check if a file is empty before attempting to read its contents

import os

filename = "example.txt"

if os.path.exists(filename) and os.path.getsize(filename) > 0:
    with open(filename, "r") as f:
        content = f.read()
        print(content)
else:
    print("File is empty or does not exist")
```

```python
# Q23 = Write a Python program that writes to a log file when an error occurs during file handling.


import logging

logging.basicConfig(filename="file_errors.log",
                    level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

try:
    # Attempt to open a non-existent file
    with open("nonexistent.txt", "r") as f:
        content = f.read()
except FileNotFoundError as e:
    logging.error("File handling error occurred: %s", e)
    print("An error occurred. Check file_errors.log for details.")
```
