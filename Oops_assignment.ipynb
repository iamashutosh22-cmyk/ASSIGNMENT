# Q1 = What is Object-Oriented Programming (OOP)..?


Object-Oriented Programming (OOP) is a programming paradigm that organizes code around classes and objects, where classes act as blueprints defining attributes and behaviors, and objects are instances that embody those definitions; it emphasizes four key principles—abstraction to hide complexity, encapsulation to protect data, inheritance to enable code reuse by creating hierarchies, and polymorphism to allow methods to take different forms—making programs more modular, reusable, and easier to maintain by modeling real-world entities and their interactions within software systems.


# Q2 = What is a class in OOP..?


In Object-Oriented Programming, a class is a blueprint or template that defines the structure and behavior of objects by specifying attributes (data fields) and methods (functions) that describe what those objects can hold and do, essentially serving as a mold from which individual instances called objects are created, allowing programmers to model real-world entities in a structured, reusable, and organized way.

# Q3 = What is an object in OOP


In Object-Oriented Programming, an object is an instance of a class that represents a specific entity with its own state and behavior, where the state is defined by the object’s attributes (data fields) and the behavior is defined by its methods (functions), making each object a concrete realization of the class blueprint that can interact with other objects to model real-world entities and their relationships within a software system.


# Q4 = What is the difference between abstraction and encapsulation..?



Abstraction focuses on what a system does by hiding unnecessary details, while encapsulation focuses on how data and methods are bundled together to protect and control access.

In Object-Oriented Programming, abstraction is about exposing only the essential features of an object and hiding the complex implementation details, allowing programmers to work with simplified models of real-world entities; it operates at the design level and helps reduce complexity by showing only relevant information. On the other hand, encapsulation is about wrapping data (attributes) and code (methods) into a single unit, typically a class, and restricting direct access to some components through access modifiers, thereby ensuring data security and integrity; it operates at the implementation level and enforces controlled interaction with the object’s internal state





# Q5 = What are dunder methods in Python



Dunder methods in Python are special built-in methods that start and end with double underscores (like __init__, __str__, or __add__), and they allow developers to define how objects of a class should behave in specific situations such as initialization, string representation, or operator overloading.

These methods are often called magic methods because they let you customize the behavior of your classes in powerful ways. For example, __init__ initializes an object when it’s created, __str__ defines how the object is displayed as a string, and __add__ lets you use the + operator with your custom objects. They’re not meant to be called directly by programmers; instead, Python automatically invokes them when certain operations occur. By implementing dunder methods, you can make your classes integrate seamlessly with Python’s syntax and built-in features, making them more intuitive and expressive to use.

```python
# Q2 = Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
#from it and implement the area() method in both.



from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

c = Circle(5)
print(c.area())

r = Rectangle(4, 6)
print(r.area())
```

```python
# Q3 =  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
# and further derive a class ElectricCar that adds a battery attribute.



class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, model):
        super().__init__(type)
        self.model = model

class ElectricCar(Car):
    def __init__(self, type, model, battery):
        super().__init__(type, model)
        self.battery = battery

v = Vehicle("Generic Vehicle")
print(v.type)

c = Car("Car", "Sedan")
print(c.type, c.model)

e = ElectricCar("Electric Car", "Tesla Model 3", "75 kWh")
print(e.type, e.model, e.battery)

```

```python
# Q5 = Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
# balance and methods to deposit, withdraw, and check balance



class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

account = BankAccount(100)
account.deposit(50)
print(account.get_balance())
account.withdraw(30)
print(account.get_balance())
```

```python
# Q7 = Create a class MathOperations with a class method add_numbers() to add two numbers and a static
#method subtract_numbers() to subtract two numbers.



class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(10, 5))
print(MathOperations.subtract_numbers(10, 5))
```

```python
# Q10 = Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#vectors.


class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Example usage
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(v3)  # Output: (6, 8)
```

```python
# Q12 = Implement a class Student with attributes name and grades. Create a method average_grade() to compute
# the average of the grades


class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

# Example usage
s1 = Student("Ashutosh", [85, 90, 78, 92])
print(s1.average_grade())  # Output: 86.25
```

```python
# Q16 = Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
#implement the sound() method.


from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"

# Example usage
cow = Cow()
sheep = Sheep()
print(cow.sound())   # Output: Moo
print(sheep.sound()) # Output: Baa
```

```python
# Q17 = Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
# returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Example usage
b1 = Book("The Alchemist", "Paulo Coelho", 1988)
print(b1.get_book_info())  # Output: 'The Alchemist' by Paulo Coelho, published in 1988
```

```python
# Q18 = Create a class House with attributes address and price. Create a derived class Mansion that adds an
#attribute number_of_rooms.


class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Example usage
h1 = House("123 Street, Varanasi", 5000000)
m1 = Mansion("456 Avenue, Delhi", 20000000, 15)

print(h1.address, h1.price)          # Output: 123 Street, Varanasi 5000000
print(m1.address, m1.price, m1.number_of_rooms)  # Output: 456 Avenue, Delhi 20000000 15
```
