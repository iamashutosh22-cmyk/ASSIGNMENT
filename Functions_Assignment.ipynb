# Theory Questions

# Q1 = 1. What is the difference between a function and a method in Python?


In Python, a function is a block of reusable code defined independently using the def keyword and can be called directly by its name, while a method is essentially a function that is associated with an object and defined within a class, meaning it automatically takes the instance (self) or class (cls) as its first parameter to operate on the data contained in that object; in short, functions are standalone, whereas methods are tied to objects and provide behavior specific to the class they belong to.

# Q2 = . Explain the concept of function arguments and parameters in Python.


In Python, parameters are the variables defined in a function’s signature that specify what kind of input the function expects, while arguments are the actual values you supply when calling the function; for instance, in def greet(name): print("Hello", name), name is the parameter, and when you call greet("Ashutosh"), the string "Ashutosh" is the argument, meaning parameters act as placeholders inside the function and arguments provide the real data that fills those placeholders during execution.





# Q3 = 3. What are the different ways to define and call a function in Python?


In Python, functions can be defined using the def keyword for regular functions, lambda for anonymous one‑line functions, or within classes as methods, and they can be called directly by name with positional arguments, keyword arguments, default arguments, or variable‑length arguments using *args and **kwargs; for example, def add(x, y): return x + y can be called as add(3, 5) with positional arguments, add(y=5, x=3) with keyword arguments, or def greet(name="Guest"): print("Hello", name) can be invoked as greet() without passing anything because of the default parameter, while lambda x: x**2 defines a quick anonymous function that can be called like f(4)

# Q4 = What is the purpose of the `return` statement in a Python function?


The return statement in a Python function is used to send a value or result back to the caller, effectively ending the function’s execution at that point and providing data that can be stored in a variable, used in expressions, or passed to other functions; without return, a function completes its task but implicitly returns None, whereas with return, you can control exactly what output the function produces, making it essential for building reusable and modular code.

# Q5 = What are iterators in Python and how do they differ from iterables


In Python, an iterable is any object capable of returning its elements one at a time (like lists, tuples, strings, or sets), while an iterator is the actual object that performs the iteration using the __next__() method; every iterator is also iterable, but not every iterable is an iterator, meaning you need to call iter() on an iterable to get its iterator, which then allows you to fetch items sequentially until exhaustion, whereas the iterable itself only provides the structure to be looped over.

# Q6 = Explain the concept of generators in Python and how they are defined.


Generators in Python are special functions that allow you to create iterators in a simple and memory‑efficient way by using the yield keyword instead of return, which means they produce values one at a time and pause execution between yields rather than generating and storing all results at once; they are defined like normal functions with def, but whenever yield is encountered, the function’s state is saved so that execution can resume from that point on the next call, making them ideal for handling large datasets or streams of data without consuming excessive memory.



# Q7 =  What are the advantages of using generators over regular functions?

Generators in Python are advantageous over regular functions because they enable lazy evaluation, producing values one at a time only when needed instead of computing and storing all results at once, which makes them highly memory‑efficient for large datasets or infinite sequences; they also preserve execution state between yields, simplifying iteration logic without requiring explicit classes or manual bookkeeping, and they allow cleaner, more concise code for tasks like streaming data, reading files line by line, or generating sequences, ultimately reducing resource usage and improving performance.



# Q8 = 8. What is a lambda function in Python and when is it typically used?


 lambda function in Python is a small anonymous function defined using the lambda keyword instead of def, and it can take any number of arguments but only contain a single expression whose result is automatically returned; these functions are typically used for short, throwaway operations where defining a full function would be unnecessary, such as in functional programming constructs like map(), filter(), and sorted() with custom keys, or for quick inline calculations and callbacks that keep code concise and readable.

# Q9 =  Explain the purpose and usage of the `map()` function in Python

The map() function in Python is used to apply a given function to each item of an iterable (like a list, tuple, or string) and returns a map object, which is an iterator that can be converted into other collections such as lists or sets; its purpose is to simplify element‑wise transformations without writing explicit loops, for example map(str.upper, ["ashutosh", "python"]) produces ["ASHUTOSH", "PYTHON"], and it is typically used when you want to perform the same operation across all elements of a sequence in a clean, functional programming style.

# Q10 = 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

In Python, map() applies a given function to each element of an iterable and returns a new iterator with the transformed results, filter() applies a function that returns a Boolean value to each element and produces an iterator containing only those elements for which the function evaluates to True, while reduce() (from the functools module) repeatedly applies a function to pairs of elements in an iterable to cumulatively reduce it to a single value; in essence, map() transforms data, filter() selects data, and reduce() aggregates data into one outcome.

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
print(powers_of_two(6))
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

tuples = [(1, 3), (4, 1), (2, 2)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])

```

```python
#9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 20, 30, 40, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

s = "Hello World"
result = ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
print(result)
```
